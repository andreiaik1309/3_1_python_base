```python
# a) Написать программу, которая удаляет из списка все элементы, стоящие на четных позициях.
def del_even_ind(arr: list):
    return [x for ind, x in enumerate(arr) if ind%2 != 0]


# b) Написать программу, которая считывает список слов и находит слова, содержащие более трех гласных букв.
def find_word(words: list):
    vowels = ['а', 'я', 'у', 'ю', 'о', 'е', 'ё', 'э', 'и', 'ы']
    def count_vowels(word: str):
        return sum([1 for x in word if x.lower() in vowels])
    return [word for word in words if count_vowels(word) > 3]


# c) Написать программу, которая находит второй по величине элемент в списке.
def second_value(arr: list):
    return sorted(arr, reverse=True)[1]


# d) Написать программу, которая удаляет из списка все дубликаты.
def del_duble(arr: list):
    return list(set(arr))


if __name__ == '__main__':
    arr = [1, 3, 5, 5, 3, 6, 7]
    print('Тестирования задания a ##############')
    print(f'Если из списка {arr} удалить все элементы с четных позиций, то получим: ', del_even_ind(arr))
    print('Тестирования задания b #############')
    words = ['Курс', 'Аргентина', 'аргентина', 'пироженное', 'печенье', 'АрбУзиНА']
    print(find_word(words))
    print('Тестирования задания c ############')
    print(f'Второй по величине элемент списка {arr} - это ', second_value(arr))
    print('Тестирования задания d ############')
    print(f'Если из списка {arr} удалить все дубли, то получим: ', del_duble(arr))
```

```python
# e) Написать программу, которая считывает данные из CSV-файла и создает словарь,
# где ключами являются значения в столбце «Name»,
# а значениями — соответствующие им словари с информацией о поле, возрасте и зарплате.
def csv_to_dict(file_name: str):
    df = pd.read_csv(file_name)
    return dict([(i,{'Gender': x,'Age': y,'Salary': z}) for i,x,y,z in zip(df['Name'], df['Gender'],df['Age'],df['Salary'])])

dict_staff = csv_to_dict('staff.csv')
print(dict_staff)
```

```python
# f) Написать программу, которая запрашивает у пользователя строку и
# выводит на экран все ее подстроки длиной не менее трех символов.
def subsent():
    sent = input('введите строку: ')
    res = []
    if len(sent) < 3:
        return 0
    for len_str in range(3, len(sent)):
        for i in range(0, len(sent) - len_str + 1):
            res.append(sent[i:i+len_str])
    res.append(sent)
    return res



subsent()
```
